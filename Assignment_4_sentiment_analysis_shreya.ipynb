```python
panctuation = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
# Removing punctuations in string
# Using loop + punctuation string
def removepanct(txt):
    for i in txt:
        if i in panctuation:
            txt = txt.replace(i, "")
    return txt
```

```python
traindf['Phrase']=traindf['Phrase'].apply(removepanct)
testdf['Phrase']=testdf['Phrase'].apply(removepanct)
```
